**Question 1.** Below we generated a dataset. Please:
1. visualize the true function $y=\sin(\pi x)$
2. on the generated training data, do linear regression to fit the data by the np.linalg.solve function (https://numpy.org/doc/stable/reference/generated/numpy.linalg.solve.html)
3. visualize the fitted function.


```python
import numpy as np
import matplotlib.pyplot as plt
import math
pi=math.pi

# generate 20 numbers from -1 to 1 with equal stepsize
x=np.linspace(-1,1,20)

# generate training target (noise contaminated!)
y=np.sin(2*pi*.5*x)+0.3*np.random.randn(x.size)

# plot the training data points
plt.plot(x, y, 'ro')

# plot the true function here
plt.plot(x, np.sin(pi*x), color="blue")
```

```python
X = [x]
X.append(np.ones(x.shape))
X = np.array(X).T
print(X.shape)

# Making the model fit to sine function
X_sin = np.ones((x.shape[0], 2))
X_sin[:,0] = np.sin(pi*x)
X_sin[:,1] = 1

# solve the regression on (X, y) and visualize the fitted function below
#X = X_sin # TRIES TO FIT TO SIN(pi*x) COMMENT OUT TO FIT MX+B
coef = np.linalg.solve(np.matmul(X.T, X), np.matmul(X.T, y))
y_model = np.matmul(X, coef)

plt.plot(x,y_model, color="red")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
pi=math.pi
# generate 10 numbers from -1 to 1 with equal stepsize
x=np.linspace(-1,1,10)

# generate training target (noise contaminated!)
y=np.sin(2*pi*.5*x)+0.4*np.random.randn(x.size)

# define a validation set
xv=np.linspace(-1,1,20)
yv=np.sin(2*pi*.5*xv)+0.4*np.random.randn(xv.size)

plt.plot(x,y,'ro', xv, yv,'gx')
```

```python
error_val= []
error_train = []


# try increasing number of basis
B = 26 # WE SET THE BASIS NUMBER HERE FOR CONVENIENCE




#for M in [1] + [2*i for i in range(1,B)]:
for M in range(1,B+1):
  plt.figure()

  basis = np.arange(M+1)
  # use BROADCASTING TO GENERATE FEATURE WITH basis so the 1st column is 1, 2nd is x, 3rd is x^2, x^3, ...
  print(x.shape)
  print(basis.shape)

  print(x[:, np.newaxis].shape)
  print(basis[np.newaxis, :])

  X = x[:, np.newaxis]**basis[np.newaxis, :]
  print(X.shape)

  # use np.linalg.solve to get the solution with/without regularization term
  reg = 0.000001



  # TO ANSWER OUR QUESTION WE NEED TO FIND NORMS OF THETA FOR THE REG WEIGHTS
  # 0, 0.0001, 0.001, 0.01, 0.1, WE WILL DO THIS FIRST BEFORE CONTINUING WITH
  # THE PROGRAM
  # WE'LL DO THIS FOR A FIXED NUMBER OF BASIS M=10 for Q2
  # For Q3 we need to check norms for basis of 1,5,10,15,20
  if M in (1,5,10,15,20):
    theta_values_q = [np.linalg.solve(np.matmul(X.T, X)+ r*np.eye(X.shape[1]), np.matmul(X.T, y))
                    for r in [0, 0.0001, 0.001, 0.01, 0.1]]
    theta_norms = np.array([np.linalg.norm(theta_val) for theta_val in theta_values_q])
    # DIVIDING THE NORMS BY THE NUMBER OF COLUMNS SQUARE ROOTED AS SUGGESTED IN Q
    theta_norms /= (X.shape[1])**(0.5)
    print(f'THETA NORMS FOR 0, 0.0001, 0.001, 0.01, 0.1 @ M = {M} => {theta_norms}')





  theta = np.linalg.solve(np.matmul(X.T, X)+ reg*np.eye(X.shape[1]), np.matmul(X.T, y))
  yhat = np.matmul(X, theta)
  # plot the fitted function
  plt.plot(x,yhat, label='fitted')

  # plot the train and validation data
  plt.plot(x,y,'ro', label='train')

  # show labels
  plt.legend()

  # compute val error and train error
  # add basis again to validation set
  Xv = xv[:, np.newaxis]**basis[np.newaxis, :]
  yhat_val = np.matmul(Xv, theta)
  error = np.sum((yv - yhat_val)**2)
  error_t = np.sum((y - yhat)**2)

  error_val.append(error)
  error_train.append(error_t)

# visualize how validation error changes as we increase basis
plt.figure()
plt.plot(error_val, label='val-err')
plt.plot(error_train, label='train-err')
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
pi=math.pi
# generate 10 numbers from -1 to 1 with equal stepsize
x=np.linspace(-1,1,10)

# generate training target (noise contaminated!)
y=np.sin(2*pi*.5*x)+0.4*np.random.randn(x.size)

# define a validation set
xv=np.linspace(-1,1,20)
yv=np.sin(2*pi*.5*xv)+0.4*np.random.randn(xv.size)

plt.plot(x,y,'ro', xv, yv,'gx')
```

```python
def get_sensi(theta, xtilde):
  yhat_orig = xtilde@theta
  yhat_p = (xtilde+np.random.normal(0., 0.2, xtilde.shape))@theta
  return np.linalg.norm(yhat_p-yhat_orig)/np.linalg.norm(yhat_orig)

np.random.seed(0)
error_val= []
error_train = []
sensi = []

# try increasing number of basis
reglist = [0.0, 0.00001, 0.0001, 0.001, 0.01, 0.1]

# CREATES EMPTY ARRAYS THAT FILL IN LOOP FOR PLOTTING AT END
senslist = []
senslistv = []


for reg in reglist:
  plt.figure()

  # always show the true function
  # plot the true function
  plt.plot(np.linspace(-1,1,50), np.sin(2*pi*.5*np.linspace(-1,1,50)), 'black', label='true')

  M = 30
  basis = np.arange(M+1)
  X = x[:, np.newaxis]**basis[np.newaxis, :]
  print(X.shape)

  # try this with different regularization weight
  useLSTQ = False # CHANGING THIS CHANGES WHETHER OUR FUNCTION USES A REGULARIZATION TERM

  theta = np.linalg.lstsq(X.T@X, X.T@y, rcond=0.0001)[0] if useLSTQ else np.linalg.solve(X.T@X + reg*np.eye(X.shape[1]), X.T@y)
  yhat = X@theta


  # WORKING OUT THE SENS VALUES AND STORING IN SENSLIST
  senslist.append(get_sensi(theta, X))





  # plot the fitted function
  plt.plot(x,yhat, label='fitted')

  # plot the train and validation data
  plt.plot(x,y,'ro', label='train')

  # show labels
  plt.legend()

  # compute val error and train error
  Xv = xv[:, np.newaxis]**basis[np.newaxis, :]

  senslistv.append(get_sensi(theta, Xv)) # DOING THE SAME FOR XVALIDATION

  yhat_val = Xv@theta
  error = np.sum((yv - yhat_val)**2)
  error_t = np.sum((y - yhat)**2)

  error_val.append(error)
  error_train.append(error_t)


# visualize how validation error changes as we increase basis
plt.figure()
plt.plot(error_val, label='val-err')
plt.plot(error_train, label='train-err')
plt.legend()



# PLOTTING SENSITIVITY VS REGULARIZATION VALUES; WE OBSERVE SENS DECREASES
# AS REG INCREASES AS EXPECTED!
plt.figure()
plt.plot(reglist, senslist)
plt.plot(reglist, senslistv)
plt.title("Regularization vs sensitivity")
```
